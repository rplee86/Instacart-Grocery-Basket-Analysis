# 2. Aggregating Variables and Deriving New Variables

#### In this Exercise, you learned how to find the aggregated mean of the “order_number” column grouped by “department_id” for a subset of your dataframe. Now, repeat this process for the entire dataframe.

```python
# Obtain mean 'order_number" value for each 'department_id'
ords_prods_merge.groupby('department_id').agg({'order_number': ['mean']})
```

#### Analyze the result. How do the results for the entire dataframe differ from those of the subset? Include your comments in a markdown cell below the executed code.

The result for the entire dataframe shows the mean order_number values for all 21 department_id's. In the subset, we only saw a the mean order_number values for 8 department_id's because those 8 department_id's were in the first 1,000,000 rows of the entire data set. Furthermore, the mean order_number values in the subset differ from the mean order_number values in the entire dataframe because the mean order_number values in the subset only pertain to the first 1,000,000 rows. In other words, the mean order_number values in the subset are sample means, not population means.

#### Follow the instructions in the Exercise for creating a loyalty flag for existing customers using the transform() and loc() functions.

#### Create a flag for cusomters with many orders using the following criteria:
1. If the maximum orders the user has made is over 40, then the customer will be labeled a “Loyal customer.”
2. If the maximum orders the user has made is over 10 but less than or equal to 40, then the customer will be labeled a “Regular customer.”
3. If the maximum orders the user has made is less than or equal to 10, then the customer will be labeled a “New customer.”

```python
# Create 'max_order' column in ords_prods_merge
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# Check the output
ords_prods_merge.head(15)
```

```python
# Check the dimensions
ords_prods_merge.shape
```

```python
# Create a 'loyalty_label' column that assigns 'Loyal customer' label
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_label'] = 'Loyal customer'
```

```python
# Create a 'loyalty_label' column that assigns 'Regular customer' label
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_label'] = 'Regular customer'
```

```python
# Create a 'loyalty_label' column that assigns 'New customer' label
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_label'] = 'New customer'
```

```python
# Check the frequencies of 'loyalty_label' column
ords_prods_merge['loyalty_label'].value_counts(dropna = False)
```

#### The team now wants to target different types of spenders in their marketing campaigns. This can be achieved by looking at the prices of the items people are buying. Create a spending flag for each user based on the average price across all their orders using the following criteria:
1. If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”
2. If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”

```python
# Create 'mean_spent' column for each user id in ords_prods_merge
ords_prods_merge['mean_spent'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Check the output
ords_prods_merge.head(15)
```

```python
# Check the dimensions
ords_prods_merge.shape
```

```python
# Create 'spender_label' column that assigns 'Low spender' label
ords_prods_merge.loc[ords_prods_merge['mean_spent'] < 10, 'spender_label'] = 'Low spender'
```

```python
# Create 'spender_label' column that assigns 'High spender' label
ords_prods_merge.loc[ords_prods_merge['mean_spent'] >= 10, 'spender_label'] = 'High spender'
```

```python
# Check the output
ords_prods_merge.head(15)
```

```python
# Check the dimensions
ords_prods_merge.shape
```

```python
# Check the frequencies of the 'spender_label' column
ords_prods_merge['spender_label'].value_counts(dropna = False)
```

#### In order to send relevant notifications to users within the app (for instance, asking users if they want to buy the same item again), the Instacart team wants you to determine frequent versus non-frequent customers. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. The criteria for the flag should be as follows:
1. If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”
2. If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”
3. If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”

```python
# Create a 'median_days_since_prior_order' column for each user id in ords_prods_merge
ords_prods_merge['median_days_since_prior_order'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# Check the output
ords_prods_merge.head(15)
```

```python
# Check the dimensions
ords_prods_merge.shape
```

```python
# Create 'customer_frequency_level' column that assigns 'Non-frequent customer' label
ords_prods_merge.loc[ords_prods_merge['median_days_since_prior_order'] > 20, 'customer_frequency_level'] = 'Non-frequent customer'
```

```python
# Create 'customer_frequency_level' column that assigns 'Regular customer' label
ords_prods_merge.loc[(ords_prods_merge['median_days_since_prior_order'] > 10) & (ords_prods_merge['median_days_since_prior_order'] <= 20), 'customer_frequency_level'] = 'Regular customer'
```

```python
# Create 'customer_frequency_level' column that assigns 'Frequent customer' label
ords_prods_merge.loc[ords_prods_merge['median_days_since_prior_order'] <= 10, 'customer_frequency_level'] = 'Frequent customer'
```

```python
# Check the output
ords_prods_merge.head(15)
```

```python
# Check the dimensions
ords_prods_merge.shape
```

```python
# Check the frequencies of 'customer_frequency_level' column
ords_prods_merge['customer_frequency_level'].value_counts(dropna = False)
```
